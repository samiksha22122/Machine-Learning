```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
centers = [[0.5,2],[-1,-1],[1.5,-1]]
```

```python
X,_ = make_blobs(n_samples=400,centers=centers,cluster_std=0.5,random_state=0)
```

```python
X
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.4,min_samples=15)
```

```python
db.fit(X)
```

```python
identified_cluster = db.fit_predict(X)
```

```python
identified_cluster
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=identified_cluster,cmap='rainbow')
plt.show()
```
