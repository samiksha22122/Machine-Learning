```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
```

```python

from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
data
```

```python
X = data.data
```

```python
X
```

```python
y = data.target
```

```python
y
```

```python
X
```

```python
X.shape
```

```python
# Feature Extraction
#The Principal Component Analysis is a popular unsupervised learning technique for reducing the dimensionality of data.
#It increases interpretability yet, at the same time, it minimizes information loss
# It helps to find the most significant features in a dataset and makes the data easy for plotting in 2D and 3D.
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
X_reduce = pca.fit_transform(X)
```

```python
X_reduce
```

```python
X_reduce.shape
```

```python

X_reduce = pd.DataFrame(X_reduce,columns=['x1','x2'])
```

```python

X_reduce
```

```python
sns.scatterplot(x='x1',y='x2',data=X_reduce,hue=y)
plt.show()
```
