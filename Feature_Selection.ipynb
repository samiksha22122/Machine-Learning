```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
data
```

```python
X = data.data
```

```python
X
```

```python
y = data.target
```

```python
y
```

```python
# removing low variance features
#Variance threshold is a simple baseline approach to feature selection.
#It removes all features whose variance doesn't meet some threshold as it is assumed that features with a higher variance may contain more useful information.
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
sel = VarianceThreshold(threshold=0.6)
```

```python
X_vt = sel.fit_transform(X)
```

```python
X_vt
```

```python
X_vt = pd.DataFrame(X_vt,columns=['x1','x2'])
```

```python
X_vt
```

```python
sns.scatterplot(x='x1',y='x2',data=X_vt,hue=y)
plt.show()
```

```python
# Select from model
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.feature_selection import SelectFromModel
```

```python
model = SVC(kernel='linear')
```

```python
model.fit(X,y)
```

```python
sfm = SelectFromModel(model,prefit=True)
```

```python
X_sfm = sfm.transform(X)
```

```python
X_sfm
```

```python
X_sfm = pd.DataFrame(X_sfm,columns=['x1','x2'])
```

```python
X_sfm
```

```python
sns.scatterplot(x='x1',y='x2',data=X_sfm,hue=y)
plt.show()
```
