```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
data
```

```python
dataset = pd.DataFrame(data.data,columns=data.feature_names)
```

```python
dataset['label'] = data.target
```

```python
dataset
```

```python
X = dataset.iloc[:,0:4].values
```

```python
X
```

```python
y = dataset.iloc[:,4].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
#GridSearchCV is a technique to search through the best parameter values from the given set of the grid of parameters.
#It is basically a cross-validation method. the model and the parameters are required to be fed in.
#Best parameter values are extracted and then the predictions are made.
```

```python
hyperparameters = {'n_estimators':[1,2,4,5,8,10,12,15,20,25,50,55,65,78,100,110,150,200]}
```

```python
model = GridSearchCV(RandomForestClassifier(),hyperparameters)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.best_params_
```

```python
# SVM
```

```python
from sklearn.svm import SVC
```

```python
hyperparameters = {'kernel':['poly','sigmoid','linear','rbf']}
```

```python
model_svm = GridSearchCV(SVC(),hyperparameters) 
```

```python
model_svm.fit(X_train,y_train)
```

```python
model_svm.score(X_test,y_test)
```

```python
model_svm.best_params_
```

```python
# KNN
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
hyperparameters = {'n_neighbors':list(range(1,200))}
```

```python
model_knn = GridSearchCV(KNeighborsClassifier(),hyperparameters)
```

```python
model_knn.fit(X_train,y_train)
```

```python
model_knn.score(X_test,y_test)
```

```python
model_knn.best_params_
```
